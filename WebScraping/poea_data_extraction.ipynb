```python
from bs4 import BeautifulSoup
import requests
page = requests.get("https://poea.gov.ph/cgi-bin/agList.asp?mode=actSB")
```

```python
page.content
```

```python
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
b = list(list(soup.children)[22].children)[2]
```

```python
b
```

```python
for child in b.children:
    print('**' ,child)
```

```python
dict = {'c_name': [], 'c_address': [], 'c_telephone': [],  'c_email': [], 'c_website': [], 'c_off_repr': [],  'c_license': []}

for b in list(list(soup.children)[22].children)[2:398]:
    dict['c_name'].append(list(b.children)[0].get_text())
    dict['c_address'].append(list(b.children)[3].string)
    dict['c_telephone'].append(list(b.children)[7].string[8:])
    c_email = list(b.children)[11].string[8:]
    if c_email == 'e':
        dict['c_email'].append('None')
    else:
        dict['c_email'].append(c_email)
    c_website = list(b.children)[15].string[8:]
    if c_website=='e':
        dict['c_website'].append('None')
    else:
        dict['c_website'].append(c_website)
    dict['c_off_repr'].append(list(b.children)[18].string)
    dict['c_license'].append(list(b.children)[24].string)
```

```python
import pandas as pd
df = pd.DataFrame(dict)
# columns=['Name', 'Address', 'Telephone'. 'Email', 'Website', 'Official representative', 'License validity']
```
